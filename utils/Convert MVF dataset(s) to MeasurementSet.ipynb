Converting the MeerKAT MVF observation data to a CASA measurement set is done using the `mvftoms.py` script provided with the `katdal` package installation.

Some general usage examples are presented, but for all available options, `mvftoms.py -h`.

For the examples below the working directory is assumed to be: `/home/jupyter/MeerKAT-Cookbook/casa`

## Example 0
Convert the full observation to an output MS, and use the `-o` option to specify an appropriate name for the output Measurement Set
```
mvftoms.py ../../example-data/1548417874/1548417874/1548417874_sdp_l0.full.rdb -o exms_full.ms
```

Default behaviour that should be noted:
* If a reference antenna is not specified, the first antenna listed will be selected by default (use `katdal` data inspection methods to help select reference antenna)
* All flags to remove suspected bad data are applied by default   
**Note: pre-flagging is generally only good to use for continuum imaging and not for spectral lines, since it will flag out the science lines.**   
For imaging the calibrators, we will select to pre-flag the data during conversion.

````casa
ls exms_full.ms
````

````casa
msfile = 'exms_full.ms'
listobs(vis=msfile)
````

````casa
target = 'PKS1934-63'
````

````casa
plotms(vis=msfile, xaxis='freq', yaxis='amp', field=target, coloraxis='baseline')
````

````casa
plotms(vis=msfile, xaxis='freq', yaxis='phase', field=target, coloraxis='baseline')
````

## Example 1

Convert only the channels defining the passband and exclude the transition band channels from the output MS
```
mvftoms.py ../../example-data/1548417874/1548417874/1548417874_sdp_l0.full.rdb -o exms_passband.ms -f -a -C '163,3885' --quack=3
```

Options specified:
* `-f` produce a full polarisation MS in CASA canonical order (HH, HV, VH, VV)
* `-a` exclude autocorrelation data
* `-C` range of frequency channels to keep
* `--quack` discard the first N dumps

````casa
msfile = 'exms_passband.ms'
target = 'PKS1934-63'
listobs(vis=msfile)
````

````casa
plotms(vis=msfile, xaxis='freq', yaxis='amp', field=target, coloraxis='baseline')
````

````casa
plotms(vis=msfile, xaxis='freq', yaxis='phase', field=target, coloraxis='baseline')
````

## Example 2

Only select the frequency range between 1.3 GHz and 1.5 GHz for convertion into MS
```
mvftoms.py ../../example-data/1548417874/1548417874/1548417874_sdp_l0.full.rdb -o exms_cleanband.ms -f -a -C '2150,3150' --quack=3 -r 'm043' --flags 'static'
```

Options specified:
* `-r` specify a reference antenna
* `--flags` remove only channels flagged in the static RFI mask

````casa
msfile = 'exms_cleanband.ms'
target = 'PKS1934-63'
listobs(vis=msfile)
````

````casa
plotms(vis=msfile, xaxis='freq', yaxis='amp', field=target, coloraxis='baseline')
````

````casa
plotms(vis=msfile, xaxis='freq', yaxis='phase', field=target, coloraxis='baseline')
````
