```casa
import numpy as np
import pyrap.tables as pt
import scipy.constants as phys

from os import F_OK
```

```casa
msfile = 'ngc3621_fixvis.ms'
splitted_ms = 'gcal_1117-248_split.ms'
print(splitted_ms)

g_cal='1117-248'

prefix = "gcal_split"
print(prefix)
```

### Apply calibration to gain calibrator and split out source

```casa
split(vis=msfile,
      field=g_cal,
      outputvis=splitted_ms,
      datacolumn='corrected')
```

```casa
visstat(splitted_ms, axis='amp', field=g_cal)
```

```casa
plotms(vis=splitted_ms,
       xaxis='time',
       yaxis='amp',
       correlation='XX,YY',
       coloraxis='corr',
       averagedata=True,
       avgbaseline=True,
       avgchannel='4096')
```

```casa
plotms(vis=splitted_ms,
       xaxis='freq',
       yaxis='amp',
       correlation='XX,YY',
       coloraxis='corr',
       averagedata=True,
       avgtime='3300',
       avgbaseline=True)
```

```casa
with pt.table(splitted_ms+'/SPECTRAL_WINDOW') as tb:
    ref_freq = tb.getcol("REF_FREQUENCY")[0]  # Hz
    print(tb.getvarcol("NUM_CHAN"))
print(ref_freq)
```
